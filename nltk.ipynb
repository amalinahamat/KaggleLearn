Read in semi-structrued text data

```python
# Read in the raw text
rawData = open(r"C:\Users\LEGION\Downloads\SMSSpamCollection.tsv").read()

# print the raw data
rawData[0:500]
```

```python
parsedData = rawData.replace('\t', '\n').split('\n')
```

```python
parsedData[0:5]
```

```python
labelList = parsedData[0::2]
textList = parsedData[1::2]
```

```python
print(labelList[0:5])
print(textList[0:5])
```

```python
import pandas as pd

fullCorpus = pd.DataFrame({
    'label' : labelList,
    'body_list' : textList
})

fullCorpus.head()
```
